# Import Required Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Loading Dataset

```python
# Load Titanic dataset from seaborn
df = sns.load_dataset('titanic')
df.head()
```

# Clean and prepare Data

```python
# Drop rows with missing age for visualization
df = df.dropna(subset=['age'])

# Fill embarked column for consistent plotting
df['embarked'] = df['embarked'].fillna('Unknown')
```

```python
df
```

# Survival Count

```python
sns.countplot(data=df, x='survived', palette = "Set2" )
plt.title('Survival Count')
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.ylabel('Passengers')
plt.show()
```

# Survival by Gender

```python
sns.countplot(data=df, x='sex', hue='survived', palette='pastel')
plt.title('Survival by Gender')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

# Age Distribution

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='age', bins=30, kde=True, color='blue')
plt.title('Age Distribution of Passengers')
plt.show()
```

# Class vs Survival Heatmap

```python
sns.heatmap(pd.crosstab(df['class'], df['survived']), annot=True, cmap='YlGnBu')
plt.title('Survival by Passenger Class')
plt.ylabel('Class')
plt.xlabel('Survived')
plt.show()
```

# Embarked vs Survival

```python
sns.countplot(data=df, x='embarked', hue='survived', palette='muted')
plt.title('Survival by Port of Embarkation')
plt.show()
```
